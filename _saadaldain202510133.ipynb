## Simple Calculator

This section defines basic arithmetic operations and demonstrates their usage.

```python
# Demonstrate the calculator functions
num1 = 10
num2 = 5

print(f"{num1} + {num2} = {add(num1, num2)}")
print(f"{num1} - {num2} = {subtract(num1, num2)}")
print(f"{num1} * {num2} = {multiply(num1, num2)}")
print(f"{num1} / {num2} = {divide(num1, num2)}")

# Test division by zero
num3 = 10
num4 = 0
print(f"{num3} / {num4} = {divide(num3, num4)}")
```

```python
def add(x, y):
    """Adds two numbers."""
    return x + y

def subtract(x, y):
    """Subtracts two numbers."""
    return x - y

def multiply(x, y):
    """Multiplies two numbers."""
    return x * y

def divide(x, y):
    """Divides two numbers. Handles division by zero."""
    if y == 0:
        return "Error: Cannot divide by zero!"
    return x / y

print("Calculator functions defined.")
```

```python
# ── Step 1: Upload your CSV file ──────────────────────────────────────────────
from google.colab import files
uploaded = files.upload()  # A file picker will appear — select HR_Analytics-Project.csv

# ── Step 2: Install / import libraries (all pre-installed in Colab) ───────────
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import io

# ── Step 3: Load dataset ──────────────────────────────────────────────────────
filename = list(uploaded.keys())[0]
df = pd.read_csv(io.BytesIO(uploaded[filename]))
print(f"Dataset loaded: {df.shape[0]} rows × {df.shape[1]} columns")

# ── Color palette ──────────────────────────────────────────────────────────────
COLORS = {
    'Research & Development': 'tab:blue',
    'Sales':                  'tab:orange',
    'Human Resources':        'tab:green',
    'Male':   'tab:cyan',
    'Female': 'tab:pink',
}

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('HR Analytics Dashboard', fontsize=16, fontweight='bold', y=1.01)

# ── 1. Attrition by Department ─────────────────────────────────────────────────
ax1 = axes[0, 0]
dept_attr = (
    df[df['Attrition'] == 'Yes']
    .groupby('Department')
    .size()
    .reset_index(name='Count')
    .sort_values('Count', ascending=False)
)
dept_colors = [COLORS[d] for d in dept_attr['Department']]
bars1 = ax1.bar(dept_attr['Department'], dept_attr['Count'],
                color=dept_colors, edgecolor='white', linewidth=0.8)
ax1.bar_label(bars1, padding=3, fontsize=10, fontweight='bold')
ax1.set_title('Attrition Count by Department', fontweight='bold')
ax1.set_xlabel('Department')
ax1.set_ylabel('Number of Employees Left')
ax1.set_ylim(0, dept_attr['Count'].max() * 1.15)
legend1 = [mpatches.Patch(color=COLORS[d], label=d) for d in dept_attr['Department']]
ax1.legend(handles=legend1, title='Department', fontsize=8)

# ── 2. Employee Count by Age Group ────────────────────────────────────────────
ax2 = axes[0, 1]
age_order = ['18-25', '26-35', '36-45', '46-55', '55+']
age_counts = df['AgeGroup'].value_counts().reindex(age_order)
age_colors = ['tab:purple', 'tab:blue', 'tab:green', 'tab:orange', 'tab:red']
bars2 = ax2.bar(age_counts.index, age_counts.values,
                color=age_colors, edgecolor='white', linewidth=0.8)
ax2.bar_label(bars2, padding=3, fontsize=10, fontweight='bold')
ax2.set_title('Employee Count by Age Group', fontweight='bold')
ax2.set_xlabel('Age Group')
ax2.set_ylabel('Number of Employees')
ax2.set_ylim(0, age_counts.max() * 1.15)
legend2 = [mpatches.Patch(color=age_colors[i], label=age_order[i]) for i in range(len(age_order))]
ax2.legend(handles=legend2, title='Age Group', fontsize=8)

# ── 3. Employee Distribution by Salary Slab ───────────────────────────────────
ax3 = axes[1, 0]
slab_order = ['Upto 5k', '5k-10k', '10k-15k', '15k+']
slab_counts = df['SalarySlab'].value_counts().reindex(slab_order)
slab_colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red']
bars3 = ax3.bar(slab_counts.index, slab_counts.values,
                color=slab_colors, edgecolor='white', linewidth=0.8)
ax3.bar_label(bars3, padding=3, fontsize=10, fontweight='bold')
ax3.set_title('Employee Distribution by Salary Slab', fontweight='bold')
ax3.set_xlabel('Salary Slab')
ax3.set_ylabel('Number of Employees')
ax3.set_ylim(0, slab_counts.max() * 1.15)
legend3 = [mpatches.Patch(color=slab_colors[i], label=slab_order[i]) for i in range(len(slab_order))]
ax3.legend(handles=legend3, title='Salary Slab', fontsize=8)

# ── 4. Top 6 Job Roles by Headcount ───────────────────────────────────────────
ax4 = axes[1, 1]
job_counts = df['JobRole'].value_counts().head(6)
role_colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown']
bars4 = ax4.bar(job_counts.index, job_counts.values,
                color=role_colors, edgecolor='white', linewidth=0.8)
ax4.bar_label(bars4, padding=3, fontsize=10, fontweight='bold')
ax4.set_title('Top 6 Job Roles by Headcount', fontweight='bold')
ax4.set_xlabel('Job Role')
ax4.set_ylabel('Number of Employees')
ax4.set_ylim(0, job_counts.max() * 1.15)
ax4.tick_params(axis='x', rotation=30)
legend4 = [mpatches.Patch(color=role_colors[i], label=job_counts.index[i]) for i in range(len(job_counts))]
ax4.legend(handles=legend4, title='Job Role', fontsize=7, loc='upper right')

plt.tight_layout()

# ── Step 4: Save and download the chart ───────────────────────────────────────
plt.savefig('hr_analytics_charts.png', dpi=150, bbox_inches='tight')
plt.show()

# Download the saved image to your local machine
files.download('hr_analytics_charts.png')
```
