```python
import pandas as pd
saad=pd.read_csv("/content/traffic.csv")
saad
```

```python
saad.dtypes
```

```python
saad.isnull().sum()
```

```python
saad["CarCount"].fillna(saad["CarCount"].mean(),inplace=True)
saad["Total"].fillna(saad["Total"].mean(),inplace=True)
saad['Traffic Situation'].fillna(saad["Traffic Situation"].mode(),inplace=True)
```

```python
saad.dropna(inplace=True)
```
