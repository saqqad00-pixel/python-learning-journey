```python
students = []
unique_majors = set()
name = input("Enter student name: ")
student_id = input("Enter student ID: ")
major = input("Enter major: ")

grades = []
grade1 = float(input("Enter grade for course 1: "))
grade2 = float(input("Enter grade for course 2: "))
grades.append(grade1)
grades.append(grade2)
total = sum(grades)
count = len(grades)
average = total / count
highest = max(grades)
lowest = min(grades)
student = {
    "name": name,
    "id": student_id,
    "major": major,
    "grades": grades,
    "average": average,
    "highest": highest
}

students.append(student)
unique_majors.add(major)
if average >= 50:
    print("Passed")
else:
    print("Failed")
    for student in students:
     print(student["major"])
    for student in students:
     print("Name:", student["name"])
    print("ID:", student["id"])
    print("Major:", student["major"])
    print("Grades:", student["grades"])
    print("Average:", student["average"])
    print("Highest:", student["highest"])
    print("Status:", "Passed" if student["average"] >= 50 else "Failed")
    print("---")
```
